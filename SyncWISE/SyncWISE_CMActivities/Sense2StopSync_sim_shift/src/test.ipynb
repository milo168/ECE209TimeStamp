```python
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
x = np.arange(40)
print(x.shape)
y = np.sin(x)
cs = CubicSpline(x, y)
x2 = np.arange(0, 40, 40/44)
fig, ax = plt.subplots(figsize=(6.5, 2))
ax.plot(x, y, 'o', label='data')
# ax.plot(x2, np.sin(x2), label='true')
ax.plot(x2, cs(x2), label="S")
```
